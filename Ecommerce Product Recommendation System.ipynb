```python
import numpy as np
import pandas as pd

class CollaborativeFiltering:
    def _init_(self, num_users, num_items, latent_features=10, learning_rate=0.001, reg_lambda=0.01, num_epochs=10):

        self.num_users = num_users
        self.num_items = num_items
        self.latent_features = latent_features
        self.learning_rate = learning_rate
        self.reg_lambda = reg_lambda
        self.num_epochs = num_epochs
        self.user_embeddings = np.random.rand(num_users, latent_features)
        self.item_embeddings = np.random.rand(num_items, latent_features)

    def fit(self, user_item_matrix):
        for epoch in range(self.num_epochs):
            for user, items in enumerate(user_item_matrix):
                for item, rating in enumerate(items):
                    if rating != 0:
                        error = rating - np.dot(self.user_embeddings[user], self.item_embeddings[item])
                        user_gradient = 2 * (error * self.item_embeddings[item] - self.reg_lambda * self.user_embeddings[user])
                        item_gradient = 2 * (error * self.user_embeddings[user] - self.reg_lambda * self.item_embeddings[item])
                        self.user_embeddings[user] += self.learning_rate * user_gradient
                        self.item_embeddings[item] += self.learning_rate * item_gradient

    def predict(self, user_id):
        return np.dot(self.user_embeddings[user_id], self.item_embeddings.T)

# Load dataset
file_path = r"C:\Users\RIO\Desktop\sample_data.csv"
ratings_df = pd.read_csv(file_path)

# Drop rows with non-numeric values in the rating column
ratings_df = ratings_df[pd.to_numeric(ratings_df['rating'], errors='coerce').notnull()]

# Mapping user and item IDs to integers
user_id_map = {user_id: idx for idx, user_id in enumerate(ratings_df['user_id'].unique())}
item_id_map = {item_id: idx for idx, item_id in enumerate(ratings_df['item_id'].unique())}

# Updating user and item IDs in the DataFrame
ratings_df['user_id'] = ratings_df['user_id'].map(user_id_map)
ratings_df['item_id'] = ratings_df['item_id'].map(item_id_map)

# Creating user-item interaction matrix
num_users = len(user_id_map)
num_items = len(item_id_map)
user_item_matrix = np.zeros((num_users, num_items))
for row in ratings_df.itertuples():
    user_item_matrix[row.user_id, row.item_id] = row.rating

# Example usage
model = CollaborativeFiltering(num_users=num_users, num_items=num_items)
model.fit(user_item_matrix)

# Recommend items for all users
recommended_items_for_all_users = {}
for user_id in range(num_users):
    predictions = model.predict(user_id)
    recommended_items = np.argsort(predictions)[::-1][:10]  # Top 10 recommended items
    recommended_items_for_all_users[user_id] = recommended_items

# Print recommended items for all users
for user_id, recommended_items in recommended_items_for_all_users.items():
    print("Recommended items for user {}:".format(user_id))
    for item_idx in recommended_items:
        original_item_id = [key for key, value in item_id_map.items() if value == item_idx][0]
        print(original_item_id)
```
